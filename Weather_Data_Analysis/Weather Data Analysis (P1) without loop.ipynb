```python
# Dictionaries....
monday = {"day": "Monday", "temperature": 23.5, "humidity": 30, "wind_speed": 5.2}
tuesday = {"day": "Tuesday", "temperature": 25.3, "humidity": 35, "wind_speed": 4.8}
wednesday = {"day": "Wednesday", "temperature": 21.8, "humidity": 35, "wind_speed": 6.1}
thursday = {"day": "Thursday", "temperature": 22.7, "humidity": 32, "wind_speed": 3.9}
friday = {"day": "Friday", "temperature": 26.1, "humidity": 40, "wind_speed": 5.6}
saturday = {"day": "Saturday", "temperature": 24.2, "humidity": 40, "wind_speed": 5.0}
sunday = {"day": "Sunday", "temperature": 25.0, "humidity": 30, "wind_speed": 4.3}

print(type(monday))

# List of dictionaries....
weather_data = [monday, tuesday, wednesday, thursday, friday, saturday, sunday]
print(type(weather_data))
print(len(weather_data))

# Temperatures....
temp1 = weather_data[0]["temperature"]
temp2 = weather_data[1]["temperature"]
temp3 = weather_data[2]["temperature"]
temp4 = weather_data[3]["temperature"]
temp5 = weather_data[4]["temperature"]
temp6 = weather_data[5]["temperature"]
temp7 = weather_data[6]["temperature"]

# Wind speeds....
wind1 = weather_data[0]["wind_speed"]
wind2 = weather_data[1]["wind_speed"]
wind3 = weather_data[2]["wind_speed"]
wind4 = weather_data[3]["wind_speed"]
wind5 = weather_data[4]["wind_speed"]
wind6 = weather_data[5]["wind_speed"]
wind7 = weather_data[6]["wind_speed"]

# Average temperature and wind speed....
average_temperature = (temp1 + temp2 + temp3 + temp4 + temp5 + temp6 + temp7) / 7
average_wind_speed = (wind1 + wind2 + wind3 + wind4 + wind5 + wind6 + wind7) / 7

print("Average Temperature:", average_temperature)
print("Average Wind Speed:", average_wind_speed)

# Use Functions....
def max_tem():
    
    # Maximum Temperature....
    if temp1 >= temp2 and temp1 >= temp3 and temp1 >= temp4 and temp1 >= temp5 and temp1 >= temp6 and temp1 >= temp7:
        max_temperature = temp1
        max_day = monday["day"]
    elif temp2 >= temp1 and temp2 >= temp3 and temp2 >= temp4 and temp2 >= temp5 and temp2 >= temp6 and temp2 >= temp7:
        max_temperature = temp2
        max_day = tuesday["day"]
    elif temp3 >= temp1 and temp3 >= temp2 and temp3 >= temp4 and temp3 >= temp5 and temp3 >= temp6 and temp3 >= temp7:
        max_temperature = temp3
        max_day = wednesday["day"]
    elif temp4 >= temp1 and temp4 >= temp2 and temp4 >= temp3 and temp4 >= temp5 and temp4 >= temp6 and temp4 >= temp7:
        max_temperature = temp4
        max_day = thursday["day"]
    elif temp5 >= temp1 and temp5 >= temp2 and temp5 >= temp3 and temp5 >= temp4 and temp5 >= temp6 and temp5 >= temp7:
        max_temperature = temp5
        max_day = friday["day"]
    elif temp6 >= temp1 and temp6 >= temp2 and temp6 >= temp3 and temp6 >= temp4 and temp6 >= temp5 and temp6 >= temp7:
        max_temperature = temp6
        max_day = saturday["day"]
    else:
        max_temperature = temp7
        max_day = sunday["day"]
    print("Maximum Temperature:", max_temperature, "on", max_day)   
    
max_tem()

def min_tem():
    
    # Minimum Temperature....
    if temp1 <= temp2 and temp1 <= temp3 and temp1 <= temp4 and temp1 <= temp5 and temp1 <= temp6 and temp1 <= temp7:
        min_temperature = temp1
        min_day = monday["day"]
    elif temp2 <= temp1 and temp2 <= temp3 and temp2 <= temp4 and temp2 <= temp5 and temp2 <= temp6 and temp2 <= temp7:
        min_temperature = temp2
        min_day = tuesday["day"]
    elif temp3 <= temp1 and temp3 <= temp2 and temp3 <= temp4 and temp3 <= temp5 and temp3 <= temp6 and temp3 <= temp7:
        min_temperature = temp3
        min_day = wednesday["day"]
    elif temp4 <= temp1 and temp4 <= temp2 and temp4 <= temp3 and temp4 <= temp5 and temp4 <= temp6 and temp4 <= temp7:
       min_temperature = temp4
       min_day = thursday["day"]
    elif temp5 <= temp1 and temp5 <= temp2 and temp5 <= temp3 and temp5 <= temp4 and temp5 <= temp6 and temp5 <= temp7:
       min_temperature = temp5
       min_day = friday["day"]
    elif temp6 <= temp1 and temp6 <= temp2 and temp6 <= temp3 and temp6 <= temp4 and temp6 <= temp5 and temp6 <= temp7:
       min_temperature = temp6
       min_day = saturday["day"]
    else:
       min_temperature = temp7
       min_day = sunday["day"]
    print("Minimum Temperature:", min_temperature, "on", min_day)    

min_tem()    

# Use Tuple....
temps_celsius = (temp1, temp2, temp3, temp4, temp5, temp6, temp7)
temps_fahrenheit = (
    temps_celsius[0] * 9/5 + 32,
    temps_celsius[1] * 9/5 + 32,
    temps_celsius[2] * 9/5 + 32,
    temps_celsius[3] * 9/5 + 32,
    temps_celsius[4] * 9/5 + 32,
    temps_celsius[5] * 9/5 + 32,
    temps_celsius[6] * 9/5 + 32,
)

print("Temperatures in Fahrenheit:", temps_fahrenheit)
print(type(temps_fahrenheit))

# Use a set....
humidity1 = weather_data[0]["humidity"]
humidity2 = weather_data[1]["humidity"]
humidity3 = weather_data[2]["humidity"]
humidity4 = weather_data[3]["humidity"]
humidity5 = weather_data[4]["humidity"]
humidity6 = weather_data[5]["humidity"]
humidity7 = weather_data[6]["humidity"]

unique_humidities = {humidity1, humidity2, humidity3, humidity4, humidity5, humidity6, humidity7}

print("Unique humidity values:", unique_humidities)
print(type(unique_humidities))

# Data type....
temperature_types = type(temp1)
humidity_types = type(humidity1)
wind_speed_type = type(wind1)
average_type = type(average_temperature)
print("Data Types - Temperature:", temperature_types, ", Humidity:", humidity_types, ", Wind Speed:", wind_speed_type, ", Average:", average_type)

# Casting....
average_temperature_rounded = int(average_temperature)
print("Rounded Average Temperature:", average_temperature_rounded)

# use copy , Clear and Del....
Days = weather_data.copy()
print(type(Days))
print(Days)

Days.clear()
print(Days)

del Days

print("Checking if a specific day has a temperature above or below or equal to 25°C.")
day_check = input("Enter the day to check (Monday to Sunday): ")
day_check = day_check.capitalize()  # Capitalize the first letter....

# Use Dictionary....
day_mapping = {
    "Monday": monday,
    "Tuesday": tuesday,
    "Wednesday": wednesday,
    "Thursday": thursday,
    "Friday": friday,
    "Saturday": saturday,
    "Sunday": sunday
}

if day_check in day_mapping:
    temp_check = day_mapping[day_check]["temperature"]
    wind_check = day_mapping[day_check]["wind_speed"]
    if temp_check > 25:
        print(f"The temperature on {day_check} is above 25°C with a wind speed of {wind_check} m/s.")
    elif temp_check < 25:
        print(f"The temperature on {day_check} is below 25°C with a wind speed of {wind_check} m/s.")
    else:
        print(f"The temperature on {day_check} is 25°C with a wind speed of {wind_check} m/s.")
else:
    print(f"{day_check} is not a valid day. Please enter a day from Monday to Sunday.")
```
