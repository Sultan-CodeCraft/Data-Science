# use polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
d = 2
```

```python
polynomial_features = PolynomialFeatures(degree=d)
```

```python
x_poly = polynomial_features.fit_transform(X)
reg2 = LinearRegression()
reg2.fit(x_poly,y)
```

```python
y_poly_pred = reg2.predict(x_poly)
y_poly_pred
```

```python
reg2.predict(polynomial_features.fit_transform([[2.5]]))
```

```python
plt.scatter(X, y, color='blue', label='Actual Salary')
plt.plot(X, y_poly_pred, color='red', linewidth=2, label=f'Predicted Salary (Polynomial Regression, Degree={d})')
plt.xlabel('Job Level')
plt.ylabel('Salary')
plt.title('Polynomial Regression: Job Level vs Salary')
plt.legend()
plt.show()
```
