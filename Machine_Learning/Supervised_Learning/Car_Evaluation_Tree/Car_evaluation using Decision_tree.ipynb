```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('car_evaluation.csv', header=None)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
cn = ['Buying', 'Maint', 'Doors', 'Persons', 'Lug_boot', 'Safety', 'Evaluation']
df.columns = cn
df
```

```python
df['Evaluation'].value_counts()
```

```python
X = df.iloc[:, :-1]  
X
```

```python
y = df.iloc[:, -1]  
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.dtypes
```

```python
import category_encoders as ce
```

```python
ec = ce.OrdinalEncoder(cols=['Buying','Maint','Doors','Persons','Lug_boot','Safety'])


X_train = ec.fit_transform(X_train)

X_test = ec.transform(X_test)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
pred
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_test,pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import classification_report
```

```python
cm = confusion_matrix(y_test,pred)
cm
```

```python
acs = accuracy_score(y_test,pred)
acs
```

```python
cr = classification_report (y_test,pred)
print(cr)
```
